```python
df.drop(columns=["Customer ID","Offer","Payment Method"],axis=1,inplace=True)
```

```python
# zip code
df_zip = pd.read_csv("telecom_zipcode_population.csv")
```

```python
# merging dataset based on 'zip code' column
df = df.merge(df_zip,on="Zip Code",how="left")
```

```python
df.drop(columns=["Zip Code"],axis=1,inplace=True)
```

```python
df.loc[df["Customer Status"] == "Stayed",['Churn Category','Churn Reason']] = 'not churned'
```

```python
fig,axes = plt.subplots(nrows=(len(num)//3),ncols=3,figsize=(16,32))
plt.subplots_adjust(wspace=0.3,hspace=0.5)
axes=axes.flatten()
for i in range(len(num)):
  sns.histplot(x=df[num[i]],hue=df["Customer Status"],palette=["red","blue"],bins=40,ax=axes[i])
```
